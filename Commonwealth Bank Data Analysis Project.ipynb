```python
import hashlib
```

```python
# Function to hash sensitive information
```

```python
def hash_value(value):
    return hashlib.sha256(str(value).encode()).hexdigest()
```

```python
# Drop sensitive columns
```

```python
anonymized_data = data_2.drop(columns=[
    'credit_card_number', 'credit_card_security_code', 'birthdate', 'address'
])
```

```python
# Hash identifiable fields
```

```python
anonymized_data['customer_id'] = anonymized_data['customer_id'].apply(hash_value)
anonymized_data['name'] = anonymized_data['name'].apply(lambda x: "Name_" + str(hash_value(x)[:6]))
anonymized_data['email'] = anonymized_data['email'].apply(lambda x: "user" + str(hash_value(x)[:6]) + "@domain.com")
```

```python
# Generalize age into ranges
```

```python
anonymized_data['age_group'] = pd.cut(anonymized_data['age'], bins=[0, 20, 30, 40, 50, 60, 70, 100], labels=[
    '0-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70+'
])
anonymized_data = anonymized_data.drop(columns=['age'])
```

```python
# Mask precise coordinates
```

```python
anonymized_data['current_location'] = anonymized_data['current_location'].apply(
    lambda x: str(eval(x)[0][:4]) + ", " + str(eval(x)[1][:4]))
```

```python
# Mask residence to city and state only
```

```python
anonymized_data['residence'] = anonymized_data['residence'].apply(
    lambda x: ', '.join(x.split('\n')[-1].split(',')[-2:])
)
```
